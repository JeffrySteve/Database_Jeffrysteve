```r
install.packages("sqldf")
install.packages("dplyr")
```

```r
library(sqldf)
library(dplyr)
```

```r
deliveries <- read.csv("deliveries.csv")
orders <- read.csv("orders.csv")
customers <- read.csv("customers.csv")
hubs <- read.csv("hubs.csv")
```

```r
colnames(deliveries)
colnames(orders)
colnames(customers)
colnames(hubs)
```

SQL QUERY 1 (DELIVERY OUTCOMES)

```r
sqldf("
SELECT delivery_status, COUNT(*) AS total
FROM deliveries
GROUP BY delivery_status
")
```

SQL QUERY 2 (HUB PERFORMANCE)

```r
sqldf("
SELECT hub_id,
       delivery_status,
       COUNT(*) AS total
FROM deliveries
GROUP BY hub_id, delivery_status
ORDER BY hub_id
")
```

SQL QUERY 3 (TOP CUSTOMERS)

```r
sqldf("
SELECT customer_id,
       COUNT(order_id) AS total_orders
FROM orders
GROUP BY customer_id
ORDER BY total_orders DESC
LIMIT 10
")
```

SQL QUERY 4 (JOIN QUERY

```r
sqldf("
SELECT d.hub_id,
       COUNT(*) AS total_deliveries
FROM deliveries d
JOIN orders o
ON d.order_id = o.order_id
JOIN customers c
ON o.customer_id = c.customer_id
GROUP BY d.hub_id
ORDER BY total_deliveries DESC
")
```

```r
install.packages("ggplot2")
library(ggplot2)
```

```r
colnames(deliveries)
```

```r
ggplot(deliveries, aes(x = delivery_status)) +
  geom_bar(fill = "steelblue") +
  labs(
    title = "Delivery Outcomes",
    x = "Delivery Status",
    y = "Count"
  )
```

```r
ggplot(deliveries,
       aes(x = factor(hub_id),
           fill = delivery_status)) +
  geom_bar(position = "stack") +
  labs(
    title = "Delivery Outcomes by Hub",
    x = "Hub ID",
    y = "Count"
  )
```
